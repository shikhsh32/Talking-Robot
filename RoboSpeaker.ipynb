```python
from gtts import gTTS
from IPython.display import Audio, display
import time
from datetime import datetime

def speak(text):
    tts = gTTS(text)
    tts.save("voice.mp3")
    return Audio("voice.mp3", autoplay=True)

def robot_response(user_input):
    user_input = user_input.lower()
    
    if "hello" in user_input or "hi" in user_input:
        return "Hello! How can I assist you today?"
    elif "your name" in user_input:
        return "I am Robot 1.1 created by Shikha Sharma."
    elif "time" in user_input:
        return f"The current time is {datetime.now().strftime('%I:%M %p')}"
    elif "date" in user_input:
        return f"Today's date is {datetime.now().strftime('%B %d, %Y')}"
    elif "joke" in user_input:
        return "Why did the computer go to therapy? Because it had too many bytes of emotional data!"
    elif "thank you" in user_input:
        return "You're welcome! Always here to help."
    elif "exit" in user_input:
        return "exit"
    else:
        return "I am still learning. Can you try asking something else?"

# Start robot
display(speak("Welcome! I am Robot 1.1 created by Shikha Sharma. What would you like to talk about?"))

while True:
    x = input("You: ")
    response = robot_response(x)
    
    if response == "exit":
        display(speak("Exiting Robot 1.1. Have a great day!"))
        break
    else:
        display(speak(response))
```
